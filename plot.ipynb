```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
ph_range = np.arange(0, 14, 0.01)
X_ph = np.array([12.83, 10.06, 9.52, 8.05, 5.95, 2.45, 1.31])
y_eme = np.array([-287.8, -143.2, -111.1, -35.5, 80.1, 275.1, 332.5])
```

```python
fig, axes = plt.subplots(1, 1, figsize=(15,7))
axes.scatter(X_ph, y_eme)
axes.set_xlabel("pH")
axes.set_ylabel("EME (mV)")
axes.set_title("EME - pH")
axes.grid(True)
axes.set_xlim(0,14)
axes.axhline(0, color='black')
axes.set_xticks(np.arange(0, 14.5, 0.5))

X_trendline = np.arange(0, 14, 0.1)
model = LinearRegression().fit(X_ph.reshape(-1, 1), y_eme.reshape(-1, 1))
y_trendline = model.predict(X_trendline.reshape(-1, 1)).reshape(1, -1)
axes.plot(X_trendline, y_trendline[0], color='r')
plt.show()
```
